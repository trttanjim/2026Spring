2) Return the first `n` Fibonacci numbers as a list. If `n < 1`, return a message.


```python
def fib(n):
    if n < 1:
        return "Invalid input"
    n_1=1
    n_2=1
    for i in range(n - 2):
       new_n_1=n_1+n_2
       n_2=n_1
       n_1=new_n_1
       
    return n_1

print(fib(5))
```

3) Implement deposit/withdraw/get_balance, then add interest in SavingsAccount.


```python
class BankAccount:
    def __init__(self, account_number, balance=0.0):
        self.account_number = account_number
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount
    
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else:
            print("Insufficient funds")
    
    def get_balance(self):
        return self.balance


class SavingsAccount(BankAccount):
    def __init__(self, account_number, balance=0.0, interest_rate=0.01):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate
    
    def apply_interest(self):
        self.balance += self.balance * self.interest_rate

# Creating a BankAccount instance
account = BankAccount("1299")
account.deposit(1500)
account.withdraw(200)
print("Balance:", account.get_balance())

# Creating a SavingsAccount instance
savings = SavingsAccount("1228", 1000, 0.06)
savings.apply_interest()
print("Savings Balance after interest:", savings.get_balance())

#note that you only need to create an __init__ for a special subclass if you are changing
# or adding a variable for the standard __init__ that is inherited from the subclass..
```

5) Dictionaries
   


```python
def word_count(s):
    counts = {}
    words = s.split()

    i = 0
    while i < len(words):
        w = words[i]
        if w in counts:
            counts[w] = counts[w] + 1
        else:
            counts[w] = 1
        i += 1

    return counts

s = "apple banana apple orange banana apple"
print(word_count(s))


#here's another version if that doesn't use split
def word_count2(s):
    counts = {}
    word = ""
    i = 0

    while i < len(s):
        if s[i] == " ":
            # we reached the end of a word
            if word != "":
                if word in counts:
                    counts[word] = counts[word] + 1
                else:
                    counts[word] = 1
                word = ""   # reset for next word
        else:
            word = word + s[i]
        i += 1

    # handle the last word (if string doesn't end with a space)
    if word != "":
        if word in counts:
            counts[word] = counts[word] + 1
        else:
            counts[word] = 1

    return counts

s = "apple banana apple orange banana apple"
print(word_count2(s))


def average_grades(grades):
    result = {}
    for name in grades:
        scores = grades[name]

        total = 0
        i = 0
        while i < len(scores):
            total = total + scores[i]
            i += 1

        avg = total / len(scores)
        result[name] = round(avg, 2)

    return result

grades = {
    "Alice": [85, 90, 88],
    "Bob": [78, 82, 80],
    "Charlie": [92, 95, 93]
}
print(average_grades(grades))


def reverse_lookup(d, value):
    keys = []
    for k in d:
        if d[k] == value:
            keys = keys + [k]
    return keys

d = {'a': 10, 'b': 20, 'c': 10, 'd': 30}
print(reverse_lookup(d, 10))


def merge_dicts(d1, d2):
    merged = {}

    for k in d1:
        merged[k] = d1[k]

    for k in d2:
        if k in merged:
            merged[k] = merged[k] + d2[k]
        else:
            merged[k] = d2[k]

    return merged

d1 = {'a': 10, 'b': 5, 'c': 8}
d2 = {'b': 7, 'c': 3, 'd': 12}
print(merge_dicts(d1, d2))
```

6) List Comprehension 


```python
# a) squares 1..10
squares = [i*i for i in range(1, 11)]
print(squares)

# b) even numbers
nums = [3, 10, 7, 8, 15, 20, 33, 12]
evens = [x for x in nums if x % 2 == 0]
print(evens)

# c) first letters
words = ["apple", "banana", "cherry", "date"]
first_letters = [w[0] for w in words]
print(first_letters)

# d) students above 80
students = {"Alice": 85, "Bob": 92, "Charlie": 78, "David": 90}
above_80 = [name for name in students if students[name] > 80]
print(above_80)

# e) flatten nested list
nested_list = [[1, 2, 3], [4, 5], [6, 7, 8]]
flat = [num for sublist in nested_list for num in sublist]
print(flat)
```
